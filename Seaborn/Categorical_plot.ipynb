Categorical Data Plot


```python
# Used to plot categorical data
# Seaborn enhance the functionality to use categorical plot
```

Import Duties

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
df

```

Boxplot

```python
sns.boxplot(x = 'tip',y= 'day',data = df)
```

Violinplot

```python
sns.violinplot(x ='day',y ='total_bill',data = df)
```

Stripplot


```python
sns.stripplot(x = 'day', y = 'total_bill',data = df)
```

Swarmplot


```python
sns.swarmplot(x = 'day', y = 'total_bill',data = df)
```

```python
sns.violinplot(x = 'tip', y = 'day', data =df)
sns.swarmplot(x = 'tip', y = 'day', data =df)
```

Barplot


```python
ax= sns.barplot(x='sex',y ='tip',data = df,estimator = np.sum )
ax.set_xlabel('Gender')
```

```python
sns.barplot(x = df['sex'],y = df['total_bill'],estimator=np.sum)
```

Countplot

```python
sns.countplot(x = df['sex'],hue = df['smoker'])
```

```python
sns.countplot(x = df['time'],hue = df['day'] )
```
