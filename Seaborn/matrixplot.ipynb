```python
# Import Duties
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

## Matrix Plot

```python
flights = sns.load_dataset('flights')
flights
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
tipscorr = tips[['total_bill','tip','size']]
tipscorr
tipscorr.corr()
```

```python
# Heat map

sns.heatmap(tipscorr.corr(),annot = True)
```

```python
# Cluster Map   

sns.clustermap(tipscorr.corr())
```

```python
# Pivot Table Heat Map

flights_pvt = flights.pivot_table(values='passengers', index ='month', columns ='year')
flights_pvt
```

```python
sns.heatmap(flights_pvt)
```
