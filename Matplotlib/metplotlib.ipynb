Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Matplotlib Basic 

```python
x = np.linspace(0,5,11)
x 
```

```python
y = x **2
y
```

```python
plt.plot(x,y)
plt.title('Hello World')
plt.xlabel('X axis')
plt.ylabel('Y axis')

```

Subplot in matplotlib

```python
plt.subplot(2,2,1)
plt.plot(x,y)
plt.subplot(2,2,2)
plt.plot(x**2,x**2)
plt.subplot(2,2,3)
plt.plot(y,x)
plt.subplot(2,2,4)
plt.plot(x,y)
```

Object Oriented Way

Figure


```python
fig = plt.figure(figsize =(6,6),dpi = 100,facecolor= 'white')
plt.plot([1,2,3], [1,4,5])
plt.show()
# plt.figure(clear = True)
```

```python
# Naming Figure

plt.figure(1,figsize=(3,3))
plt.plot([1,2,3], [1,4,9])

plt.figure(2,figsize=(3,3))
plt.plot([1,2,3], [1,2,3])

plt.show()
```

```python
fig = plt.figure(figsize=(5,5),dpi = 100)
axis1 = fig.add_axes([0.1,0.1,0.8,0.8])
axis2 = fig.add_axes([0.2,0.58,0.3,0.3])
axis3 = fig.add_axes([0.58,0.58,0.3,0.3])

axis1.plot(x,y)
axis2.plot(x,y)
axis3.plot(y,y)
```

```python
fig = plt.figure(figsize=(6,4))
ax = fig.add_subplot(111)

ax.plot([1,2,3], [1,4,9])
```

Saving plots

```python
fig = plt.figure()
ax =fig.add_axes((0.1,0.1,0.8,0.8))
ax.plot(x,y)
plt.savefig('basicplot.png')
```
