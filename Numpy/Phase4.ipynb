```python
import numpy as np
```

```python
arr = np.array([1,2,np.nan,4,np.nan,6,7,np.nan,9,10]).reshape(2,5)
print("Original Array with NaNs:\n", arr)

print("\nChecking nan value : \n", np.isnan(arr))

cleaned_arr = arr[~np.isnan(arr)]
print("\nCleaned Array (NaNs removed):\n", cleaned_arr)

replaced_arr = np.where(np.isnan(arr), 10, arr)
print("\nArray with NaNs replaced by 0:\n", replaced_arr)
```

```python
arr = np.array([1,2,3,np.inf,5,6,-np.inf,8,9,10])
print("\nOriginal Array with Infs:\n", arr)

cleaned = arr[~np.isinf(arr)]
print("\nCleaned Array (Infs removed):\n", cleaned)

replaced = np.where(np.isinf(arr),1000, arr)
print("\nArray with Infs replaced by 0:\n", replaced)
```
